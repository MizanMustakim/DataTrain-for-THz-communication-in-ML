```python
import pandas as pd
import numpy as np
df = pd.read_excel(r'coefficient2.xlsx')
```

# Distance Relation Function

```python
frequency = df.loc[1:,'Unnamed: 0']
skin_refractive = df.loc[1:,'skin']
skin_absorption = df.loc[1:,'skin.1']
fat_refractive = df.loc[1:,'Fat']
fat_absorption = df.loc[1:,'Fat.1']

M= []
def DistanceRelation(index):
    distance = np.linspace(.1*10**(-3), 1*10**(-3), 20)
    s_relation = []
    f_relation = []
    

    for i in range(len(distance)):
    #     skin
        skin_spreading = ((4*np.pi*(frequency[index]*10**12) * skin_refractive[index] * (distance[i])) / (3 * 10**8)) ** 2
        skin_absolute = np.exp((skin_absorption[index] * 10**3) * (distance[i]))
        skin_total = skin_spreading * skin_absolute
        skin_paLo = 10 * np.log10(skin_total)
        s_relation.append(skin_paLo)

    #     fat
        fat_spreading = ((4*np.pi*(frequency[index]*10**12) * fat_refractive[index] * distance[i]) / (3 * 10**8)) ** 2
        fat_absolute = np.exp((fat_absorption[index] * 10**3) * distance[i])
        fat_total = fat_spreading * fat_absolute
        fat_paLo = 10 * np.log10(fat_total)        
        f_relation.append(fat_paLo)


    for i in range(len(s_relation)):
        for j in range(len(f_relation)):
            if abs(s_relation[i] - f_relation[j]) < 1:
                s_index = np.where(distance==distance[i]) 
                f_index = np.where(distance==distance[j])
    # print(s_index,f_index)

    expo = np.sqrt(np.exp(skin_absorption[index]*10**(3) * distance[s_index]) / np.exp(fat_absorption[index]*10**(3) * distance[f_index]))
    x = expo * (skin_refractive[index] / fat_refractive[index])
    M.append(*x)

    s_relation.clear()
    f_relation.clear()
    
```

```python
for i in range(len(frequency)):
    (DistanceRelation(i+1))

Data = {'Frequency': frequency,
       'Skin Refractive Index' : skin_refractive,
       'Skin Absorption Coefficient (mm^-1)': skin_absorption,
       'Fat Refractive Index': fat_refractive,
       'Fat Absorption Coefficient (mm^-1)' : fat_absorption}
df1 = pd.DataFrame(Data, index=range(1,11))
df1['The distance relation of each frequency between "SKIN" & "FAT"'] = M
M.clear()
df1.to_csv('Update 3.csv')
df1
```
